```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Employee": [
        "Ali", "Sara", "Ahmed", "Ayesha", "Usman",
        "Fatima", "Hamza", "Zain", "Noor", "Hina",
        "Bilal", "Maryam"
    ],

    "Experience": [1, 2, 3, 4, 5, 6, 2, 7, 8, 4, 9, 5],

    "Salary": [35000, 42000, 50000, 58000, 65000, 72000,
               45000, 80000, 90000, 60000, 95000, 68000],

    "Department": [
        "HR", "IT", "IT", "Finance", "HR", "IT",
        "Finance", "IT", "HR", "Finance", "IT", "HR"
    ]
}

df = pd.DataFrame(data)

plt.scatter(
    df['Experience'],df['Salary'],marker='o',s=70,alpha=0.7,cmap='viridis'
)

plt.annotate(
    "Highest Salary Employee",
    xy=(9,95000),
    xytext=(5,95000),
    arrowprops=dict(arrowstyle='->')

)

plt.title('Salary vs Experience')
plt.xlabel('Experience')
plt.ylabel("Salary")
plt.grid(True,alpha=0.6)
plt.show()
```

```python
labels = ['Rent', 'Food', 'Savings', 'Entertainment']
sizes = [40, 25, 20, 15]
colors = ['#4C72B0', '#DD8452', '#55A868', '#C44E52']
explode = [0.05, 0, 0.1, 0]   # Savings slice highlight

plt.figure(figsize=(6,6))

plt.pie(sizes, labels=labels, colors=colors, explode=explode,
        autopct='%1.1f%%',
        shadow=True,
        startangle=90,
        wedgeprops={'edgecolor':'white'},
        textprops={'weight':'bold'})

plt.title('Monthly Budget Breakdown',weight='bold')

plt.axis('equal')   # circle perfectly round rakhne ke liye

plt.legend(labels,title='Category',bbox_to_anchor=(1,1))
plt.show()
```
